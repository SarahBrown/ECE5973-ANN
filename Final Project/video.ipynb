```python
#required imports
import pandas as pd
import numpy as np
import tensorflow as tf
import cv2 as cv2

from tensorflow.keras import datasets, layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import os
import imutils

from matplotlib import pyplot as plt
%matplotlib inline
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
#set text
font = cv2.FONT_HERSHEY_SIMPLEX     
fontScale = 1.0
color = (255, 0, 0)
thickness = 2
```

```python
model = models.load_model("model3.h5")
img_generator = ImageDataGenerator(rescale=1/255.0, rotation_range = 0)
```

```python
def processImg(img, img_size):
    img = cv2.resize(img, (720,int(1080*(1080/1920))))
    img = img[150:350,200:400]
    img_size_h = int(img_size*img.shape[0]/img.shape[1])
    img = cv2.resize(img,(img_size, img_size_h))
    return img
```

```python
def processImgOtherView(img, img_size):
    img = cv2.resize(img, (720,int(1080*(1080/1920))))
    img = img[100:200,450:550]
    img_size_h = int(img_size*img.shape[0]/img.shape[1])
    img = cv2.resize(img,(img_size, img_size_h))
    return img
```

```python
def showVideoPred(cap, window_name, out_name):
    cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)
    cv2.setWindowProperty(window_name,cv2.WND_PROP_FULLSCREEN,cv2.WINDOW_FULLSCREEN) #to force window to be on top
    cv2.setWindowProperty(window_name,cv2.WND_PROP_FULLSCREEN,cv2.WINDOW_NORMAL) #to force window to be on top
    cv2.resizeWindow(window_name, 960, 640) 
    
    fourcc =  cv2.VideoWriter_fourcc(*'XVID')
        
    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))
   
    size = (frame_width, frame_height)
   
    result = cv2.VideoWriter(out_name, 
                         cv2.VideoWriter_fourcc(*'MJPG'),
                         20, size)
    while (True):
        ret, frame = cap.read()
        if (ret == True):
            frames = []
            #process_frame = processImg(frame, 64)
            process_frame = processImgOtherView(frame, 64)
            frames.append(process_frame)
            frame_np = np.array(frames)
            frame_flow = img_generator.flow(frame_np)
            frame_pred = model.predict(frame_flow, verbose = 1)
            prob = frame_pred[0][0]
            print(prob)
            if (prob > 0.5):
                mail = "Mail has arrived."
            else:
                mail = "Mailtruck not at mailbox."

            frame = cv2.putText(frame, str(prob), (100,100), font, fontScale, color, thickness, cv2.LINE_AA)
            frame = cv2.putText(frame, mail, (100,150), font, fontScale, color, thickness, cv2.LINE_AA)
            result.write(frame)
            cv2.imshow(window_name,frame)
        else:
            cv2.waitKey(1)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    result.release()
    cap.release()
    cv2.destroyAllWindows()
```
